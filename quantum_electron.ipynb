```python
N = 100
n = 1
xs = np.linspace(0, 1, N)
psi_real = np.sqrt(2) * np.sin(np.pi * n * xs)
psi_imaginary = np.zeros((N))

def compute_H_r(psi_real):
    H_r = np.zeros(N)
    H_r[1:-1] = 0.5 * (psi_real[1:-1] * 2 - psi_real[:-2] - psi_real[2:]) / delta_x ** 2 + kappa * psi_real[1:-1] * np.sin(omega * tau)
    return H_r

def compute_H_i(psi_imaginary):
    H_i = np.zeros(N)
    H_i[1:-1] = 0.5 * (psi_imaginary[1:-1] * 2 - psi_imaginary[:-2] - psi_imaginary[2:]) / delta_x ** 2 + kappa * psi_imaginary[1:-1] * np.sin(omega * tau)
    return H_i
```

```python
class Simulation(object):
    def __init__(self, tau=0, n=1, delta_tau=0.00005):
        self.N = 100
        self.n = n
        self.tau = tau
        self.delta_tau = delta_tau
        
        self.kappa = 0
        self.omega = 0
        self.xs = np.linspace(0, 1, self.N)
        
        self.delta_x = self.xs[1] - self.xs[0]
        
        self.psi_real = np.sqrt(2) * np.sin(np.pi * self.n * xs)
        self.psi_imaginary = np.zeros((self.N))
        self.H_r = self.compute_H_r(self.psi_real)
        self.H_i = self.compute_H_i(self.psi_imaginary)

    def step(self):
        self.tau += self.delta_tau
        psi_real_half = self.psi_real + self.H_i * self.delta_tau / 2
        self.H_r = self.compute_H_r(psi_real_half)
        self.psi_imaginary = self.psi_imaginary - self.H_r * self.delta_tau
        
        self.H_i = self.compute_H_i(self.psi_imaginary)
        self.psi_real = psi_real_half + self.H_i * self.delta_tau / 2
        return self.psi_real, self.psi_imaginary
    
    def compute_system_parameters(self):
        N = self.delta_x * (np.dot(self.psi_imaginary, self.psi_imaginary) + np.dot(self.psi_real, self.psi_real))
        x = self.delta_x * np.dot(self.xs, (np.power(self.psi_imaginary, 2) + np.power(self.psi_real, 2)))
        
        epsilon = self.delta_x * (np.dot(self.psi_imaginary, H_i) + np.dot(self.psi_real, H_r))
        ro = np.dot(self.psi_imaginary, self.psi_imaginary) + np.dot(self.psi_real, self.psi_real)
        
        return self.tau, N, x, epsilon, ro
        
    def compute_H_r(self, psi_real):
        H_r = np.zeros(self.N)
        H_r[1:-1] = 0.5 * (psi_real[1:-1] * 2 - psi_real[:-2] - psi_real[2:]) / self.delta_x ** 2 + self.kappa * psi_real[1:-1] * np.sin(self.omega * self.tau)
        return H_r

    def compute_H_i(self, psi_imaginary):
        H_i = np.zeros(self.N)
        H_i[1:-1] = 0.5 * (psi_imaginary[1:-1] * 2 - psi_imaginary[:-2] - psi_imaginary[2:]) / self.delta_x ** 2 + self.kappa * psi_imaginary[1:-1] * np.sin(self.omega * self.tau)
        return H_i
```

```python
def run_simulation():
    reals_all = []
    imaginaries_all = []
    taus_all = []
    ns_all = []
    x_means_all = []
    epsilons_all = []
    ros_all = []
    levels = [1, 4, 9]
    for n in levels:
        reals = []
        imaginaries = []
        taus = []
        ns = []
        x_means = []
        epsilons = []
        ros = []
        simulation = Simulation(n=n)
        for y in range(40):
            for x in range(4000):
                psi_real, psi_imaginary = simulation.step()
            t, N, x, epsilon, ro = simulation.compute_system_parameters()
            taus.append(t)
            ns.append(n)
            x_means.append(x)
            epsilons.append(epsilon)
            ros.append(ro)
            reals.append(psi_real)
            imaginaries.append(psi_imaginary)

        reals_all.append(reals)
        imaginaries_all.append(imaginaries)
        taus_all.append(taus)
        ns_all.append(ns)
        x_means_all.append(x_means)
        epsilons_all.append(epsilons)
        ros_all.append(ros)
    
    return levels, xs, (reals_all, imaginaries_all, taus_all,
                         ns_all, x_means_all, epsilons_all, ros_all)

def plot_simulations(levels, xs,results):
    for i, n in enumerate(levels):
        plot_simulation( n, xs, *[result[i] for result in results])

        
def plot_simulation(n, xs, reals, imaginaries, taus, ns, x_means, epsilons, ros):
    plt.figure()
    plt.title(r'Real part of $\Psi_{%s}$' % n)
    for psi in reals:
        plt.plot(xs, psi)

    plt.figure()
    plt.title(r'Imaginary part of $\Psi_{%s}$' % n)
    for psi in imaginaries:
        plt.plot(xs, psi)
    plt.figure()
    plt.plot(taus, ns)
    plt.title('N(t)')
    plt.figure()
    plt.plot(taus, x_means)
    plt.title('x_mean(t)')
    plt.figure()
    plt.plot(taus, epsilons)
    plt.title(r'$\epsilon(t)$')
    plt.figure()
    plt.plot(taus, ros)
    plt.title(r'$\rho(t)$')
```

```python
for delta in [0.0001, 0.00005, 0.00001]:
    reals = []
    imaginaries = []
    taus = []
    ns = []
    x_means = []
    epsilons = []
    ros = []
    simulation = Simulation(n=1, delta_tau=delta)
    for y in range(50):
        for x in range(4000):
            psi_real, psi_imaginary = simulation.step()
        t, N, x, epsilon, ro = simulation.compute_system_parameters()
        taus.append(t)
        ns.append(n)
        x_means.append(x)
        epsilons.append(epsilon)
        ros.append(ro)
        reals.append(psi_real)
        imaginaries.append(psi_imaginary)
        
    plot_simulation(1, xs, reals, imaginaries, taus, ns, x_means, epsilons, ros)
```

```python
levels, xs, results = run_simulation()
```

```python
plot_simulations(levels, xs, results)
```
