## Analiza symulacji dynamiki kwantowego elektronu

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
def load_from_file(filename):
    xs = []
    ys = []
    with open(filename) as f:
        for line in f:
            x, y = line.split()
            xs.append(float(x))
            ys.append(float(y))
    return xs, ys
```

## 1. Badanie stabilności algorytmu

Przy dużych wartościach $\tau$ takich jak np. 0.001 algorytm jest bardzo niestabilny.

Dla wartości $\tau$ 0.0001 algorytm jest już całkowicie stabilny.

```python
import os
```

```python
# small function to visualize parameters of simulation
def visualize_parameters(basedir, output_suffix, name_filter=None, full_title=False):
    filenames = [basedir + x for x in os.listdir(basedir) if output_suffix in x and 'psi_im' not in x and 'psi_re' not in x]
    if name_filter:
        filenames = [name for name in filenames if name_filter in name]
        
    for filename in filenames:
        xs, ys = load_from_file(filename)
        plt.figure(figsize=(10, 8))
        plt.plot(xs, ys)
        if not full_title:
            title = filename.split('/')[1].split('_')[0]
        else:
            title = filename.split('/')[1].split('output')[0].replace('_', ' ')
        plt.title(title)
        plt.xlabel(r'$\tau [ps]$')
        plt.show()
```

Czy algorytm jest stabilny? Jak wyglądają jego parametry w trakcie symulacji?

## 3. Wyliczanie Krzywej rezonansowej

```python
visualize_parameters('kappa/', 'percents3.dat', name_filter='epsilon', full_title=True);
```

```python
def max_epsilons(basedir, output_suffix, name_filter=None, title=False):
    filenames = [basedir + x for x in os.listdir(basedir) if output_suffix in x and 'epsilon' in x]
    if name_filter:
        filenames = [name for name in filenames if name_filter in name]
    
    omegas = []
    epsilons = []
    for filename in filenames:
        xs, ys = load_from_file(filename)
        epsilons.append(max(ys))
        omega = float(filename.split('/')[1].split('output')[0].split('_')[-1])
        omegas.append(omega)
    
    plt.figure(figsize=(10, 8))
    plt.plot(omegas, epsilons, 'o')
    plt.title(title)
    plt.xlabel(r'$\tau [ps]$')
    plt.show()
```

# Krzywa rezonansowa wokół pierwszej częstości

```python
max_epsilons('kappa/', 'percents4.dat', name_filter='epsilon', title=r"Max $\epsilon (\omega)$");
```

# Krzywa rezonansowa wokół drugiej częstości

```python
max_epsilons('kappa/', 'percents3.dat', name_filter='epsilon', title=r"Max $\epsilon (\omega)$");
```
